```python
clustering_data['name'] = data_with_names['name']
clustering_data['x'] = data_emb['x']
clustering_data['y'] = data_emb['y']
clustering_data['z'] = data_emb['z']
clustering_data['cluster'] = data_with_names['topic']
```
